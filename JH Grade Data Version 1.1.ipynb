# Analysis of Longitudinal Grade Data for 7th and 8th Grade Courses

This document will analyze grade data in junior high courses from 2006-2019.  The fall semester has been chosen as a guide to keep the data set manageable. Even so, the dataset contains 23,492 lines. Each line of data contains a single grade entry for one student in one class.  Names and other student details have been stripped from the dataset, leaving behind only aggregate grade data with no identifying information.

#### Importing Necessary Libraries
Pandas is a Python library which contains tools for reading and manipulating .csv or .xlsx files.  Numpy contains math functions, and mathplotlib provides tools for graphing data.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Methodology

This analysis will rely on using Pandas to search the extensive data set for grade entries related to a particular course, take all of the thousands of entries for that course, and create averages by year, and then return the results in a table, or dataframe in Pandas parlance. An example of such a table for 7th grade history is below. A similar dataframe exists for every course analyzed in this document. Further down, these dataframes will not be printed, but will exist in the backbround, providing the data for the charts.  

```python
grades = pd.read_csv('jhgrades.csv')

history7 = grades.loc[(grades['Course Name'] == 'History 7')]
history7sort=history7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})

print (history7sort)
```

#### Analyzing the Four Year-Long, Core 7th Grade Courses

The chart below will display the average course grade by year for seventh-grade History, Prealgebra, Science, and English.  

```python
grades = pd.read_csv('jhgrades.csv')

history7 = grades.loc[(grades['Course Name'] == 'History 7')]
prealgebra = grades.loc[(grades['Course Name'] == 'Pre-Algebra')]
science7 = grades.loc[(grades['Course Name'] == 'Science 7')]
english7 = grades.loc[(grades['Course Name'] == 'English 7')]


history7sort=history7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
prealgebrasort=prealgebra.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
science7sort=science7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
english7sort=english7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})

plt.figure(figsize=(13,8))

plt.title('Averages in Year-Long Required 7th Grade Courses', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(history7sort.Year, history7sort.Percent,'b.-', label='History 7', color='green')
plt.plot(prealgebrasort.Year, prealgebrasort.Percent, 'b.-', label='Prealgebra', color='blue')
plt.plot(science7sort.Year, science7sort.Percent, 'b.-', label='Science 7', color='red')
plt.plot(english7sort.Year, english7sort.Percent, 'b.-', label='English 7', color='orange')

plt.xlabel('Year')
plt.ylabel('Average Grade')

plt.legend()

plt.savefig('History7sort.png', dpi=300)

plt.show()


```

#### Analyzing One-Semester Required 7th Grade Courses

This chart does a similar analysis for one-semester required classes, including Art, Music, P.E., and Computer Essentials

```python
grades = pd.read_csv('jhgrades.csv')

coreart7 = grades.loc[(grades['Course Name'] == 'CoreArt History/Appreciation 7')]
coremusic7 = grades.loc[(grades['Course Name'] == 'Core Music History/Theory 7')]
corepe7 = grades.loc[(grades['Course Name'] == 'P.E. 7')]
computeressentials = grades.loc[(grades['Course Name'] == 'Computer Essentials')]


coreart7sort=coreart7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
coremusic7sort=coremusic7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
corepe7sort=corepe7.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
computeressentialssort=computeressentials.groupby(['Year'],as_index=False).agg({'Percent':'mean'})

plt.figure(figsize=(13,8))

plt.title('Averages in One-Semester Required 7th Grade Courses', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(coreart7sort.Year, coreart7sort.Percent,'b.-', label='CoreArt 7', color='green')
plt.plot(coremusic7sort.Year, coremusic7sort.Percent, 'b.-', label='Core Music 7', color='blue')
plt.plot(corepe7sort.Year, corepe7sort.Percent, 'b.-', label='P.E. 7', color='red')
plt.plot(computeressentialssort.Year, computeressentialssort.Percent, 'b.-', label='Computer Essentials', color='orange')

plt.xlabel('Year')
plt.ylabel('Average Grade')

plt.legend()

plt.savefig('History7sort.png', dpi=300)

plt.show()
```

#### Analyzing the Four Year-Long, Core 8th Grade Courses

Now the equivalent 8th grade year-long classes are graphed.

```python
grades = pd.read_csv('jhgrades.csv')

algebra1 = grades.loc[(grades['Course Name'] == 'Algebra 1 (H)')]
english8 = grades.loc[(grades['Course Name'] == 'English 8')]
science8 = grades.loc[(grades['Course Name'] == 'Science 8')]
history8 = grades.loc[(grades['Course Name'] == 'History 8')]


algebra1sort=algebra1.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
english8sort=english8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
science8sort=science8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
history8sort=history8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})


plt.figure(figsize=(13,8))

plt.title('Averages in Year-Long Required 8th Grade Courses', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(algebra1sort.Year, algebra1sort.Percent,'b.-', label='Algebra 1', color='purple')
plt.plot(english8sort.Year, english8sort.Percent,'b.-', label='English 8', color='Green')
plt.plot(science8sort.Year, science8sort.Percent,'b.-', label='Science 8', color='red')
plt.plot(history8sort.Year, history8sort.Percent,'b.-', label='History 8', color='blue')


plt.xlabel('Year')
plt.ylabel('Average Grade')

plt.legend()

plt.savefig('8thgradecoresort.png', dpi=300)

plt.show()
```

#### Analyzing One-Semester Required 8th Grade Courses

Finally, 8th grade one-semester, required classes are considered. 

```python
grades = pd.read_csv('jhgrades.csv')

coreart8 = grades.loc[(grades['Course Name'] == 'CoreArt History/Appreciation 8')]
coremusic8 = grades.loc[(grades['Course Name'] == 'Core Music History/Theory 8-MF')]
corepe8 = grades.loc[(grades['Course Name'] == 'P.E. 8 T-W-TH')]
composition1 = grades.loc[(grades['Course Name'] == 'Composition 1')]


coreart8sort=coreart8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
coremusic8sort=coremusic8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
corepe8sort=corepe8.groupby(['Year'],as_index=False).agg({'Percent':'mean'})
composition1sort=composition1.groupby(['Year'],as_index=False).agg({'Percent':'mean'})


plt.figure(figsize=(13,8))

plt.title('Averages in One-Semeseter Required 7th Grade Courses', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(coreart8sort.Year, coreart8sort.Percent,'b.-', label='CoreArt 8', color='purple')
plt.plot(coremusic8sort.Year, coremusic8sort.Percent,'b.-', label='CoreMusic 8', color='Green')
plt.plot(corepe8sort.Year, corepe8sort.Percent,'b.-', label='CorePE 8', color='red')
plt.plot(composition1sort.Year, composition1sort.Percent,'b.-', label='Composition 1', color='blue')


plt.xlabel('Year')
plt.ylabel('Average Grade')

plt.legend()

plt.savefig('8thgradecoresort.png', dpi=300)

plt.show()

```

#### Final Thoughts

One of my key questions in analyzing this data was whether there had been wide-spread grade inflation over time.  It seems that the answer is no.  Some individual classes have exprienced grade inflation, but in general the data reveals many courses where average grades are flat or declining. 

It is interesting to consider the impact of individual teachers.  In some cases, significant movement in the averages seems obviously linked to the arrival or departure of personnel.  That overall trend strikes me as inevitable, but the particular examples can be interesting.  

The data also highlights which courses are habitually more difficult for students than others.  In 7th grade, for example, averages in Computer Essentials are consistently strikingly lower than other classes, and that trend has been consistent across both recent instructors in the class.  In both 7th and 8th grade History has a consistently high average. That trend, too, continues across multiple instructors.
